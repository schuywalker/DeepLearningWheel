```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
combined_dfs = pd.read_csv('Models/results.csv')
```

```python
combined_dfs.shape
batches = []
for i in range(0, combined_dfs.shape[0], 50):
    batches.append(combined_dfs[i:i+50])
```

```python
batch = batches[0]
input_data = batch[0:30]['close']
ground_truth = batch[30:40]['close']
predictions = batch[40:50]['close']
```

```python
# Plotting
whole_truth = batch[0:40]['close']
predictions = batch[40:50]['close']
time_input = np.arange(40)
time_predict = np.arange(30, 40)

x_input = range(0, len(input_data))
x_predictions = range(len(input_data), len(input_data) + len(predictions))
x_ground_truth = range(len(input_data) , len(input_data) + len(predictions))


plt.figure(figsize=(10,5))
plt.scatter(time_input, whole_truth, label='Input Data', color='blue', marker='o')
plt.plot(x_ground_truth, ground_truth, label='Ground Truth', color='green', linestyle='-.')
plt.plot(x_predictions, predictions, label='Predictions', color='green', linestyle='--')
plt.scatter(time_predict, predictions, label='Predictions', color='red', marker='x')
plt.plot()
# plt.scatter(time_truth, ground_truth, label='Ground Truth', color='red', marker='s')
plt.legend()
plt.title("Close Prices over Time")
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5 * len(batches)))  # Adjusting the height based on number of batches

for idx, batch in enumerate(batches):
    input_data = batch[0:30]['close']
    ground_truth = batch[30:40]['close']
    predictions = batch[40:50]['close']

    time_input = np.arange(30)
    time_predict = np.arange(30, 40)
    time_truth = np.arange(30, 40)

    plt.subplot(len(batches), 1, idx + 1)
    
    plt.scatter(time_input, input_data, label='Input Data', color='blue', marker='o')
    plt.scatter(time_predict, predictions, label='Predictions', color='green', marker='x')
    plt.scatter(time_truth, ground_truth, label='Ground Truth', color='blue', marker='s')
    
    plt.legend()
    plt.title(f"Batch {idx + 1}: Close Prices over Time")
    plt.xlabel("Time")
    plt.ylabel("Close Price")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
